## $$
\partial_t u + u \partial_x u = \nu \partial^{2}_{xx} u
$$

- $u\partial_x u$: convective term
- $\nu \partial_{xx}^{2}u$: diffusive term

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("bmh")
```

# TVD : Total Variation Diminishing (Harten, 1983) 

```python
Nx= 200
#cfl= 0.2
tmax = 1 # time 0 ~ 1

#viscosity_coeff = (0.01/np.pi)
viscosity_coeff = 0.02

x = np.linspace(-1,1, 200)

#t = 0  , I.C
u = -1*np.sin(np.pi * x)
dx = abs(x[1]-x[0])

dt = 0.002# cfl * dx / np.max(np.abs(u))

nt = int(tmax/ dt) #+ 1

uf = np.zeros((nt,Nx))
uf[0, :] = u

print(nt)
print(dt)
print(nt * dt, tmax )
```

```python
def f(u):
    y = 0.5 * u**2
    yp = u
    return y, yp
```

```python
def minmod(a,b):
    return 0.5 * (np.sign(a)+ np.sign(b)) * np.minimum(np.abs(a), np.abs(b))
```

```python
def RHS(u, dx, viscosity_coeff):
    #diffusion term
    diffusion_term = viscosity_coeff * (np.roll(u,1)- 2*u + np.roll(u,-1))/ dx**2)
    
    ux = minmod((u - np.roll(u,1))/dx ,  (np.roll(u,-1) - u)/dx)
    
    uL = np.roll(u -0.5 * dx*ux,1)
    uR = u -0.5 * dx*ux
    fL,fpL = f(uL)
    fR,fpR = f(uR)
    a = np.maximum(np.abs(fpL), np.abs(fpR))
    
    H =0.5 * (fL + fR - a * (uR - uL))
    
    conv_term  = -(np.roll(H,-1)-H)/dx
    
    y = conv_term + diffusion_term
    return y
```

```python
for i in range(1, nt):
    u1 = u + dt * RHS(u,dx,viscosity_coeff)
    u = 0.5 * u + 0.5 * (u1 + dt * RHS(u1, dx,viscosity_coeff))
    uf[i, :] = u
```

```python
plt.figure(figsize=(8,6));
plt.plot(x, uf[0], '-o', label="Initial Wave");
plt.plot(x, uf[-1], '-o', label="Final Wave");
plt.xlabel(r'$x$ [m]', fontsize=16);
plt.ylabel(r"$u$ [m/s]", fontsize=16);
```

# Inverse Physics-Informed Neural Networks (I-PINNs)

## Preparing training data

```python
uf # all u from num.sim.
tf = np.linspace(0, 1, nt) # t 
xf = x # x
```

```python
tf_tensor = torch.tensor(tf)
xf_tensor = torch.tensor(xf)
print(len(tf_tensor))
print(len(xf_tensor))
```

```python
combined_tensor_x_train = torch.empty((len(tf)*len(xf), 2), dtype=torch.float32)
```

```python
index = 0
for i in range(len(tf)):
    for j in range(len(xf)):
        combined_tensor_x_train[index][0] = xf_tensor[j]
        combined_tensor_x_train[index][1] = tf_tensor[i]
        index+=1
```

```python
y_train = torch.tensor(uf, dtype=torch.float32) # from tvd
y_train = y_train.view(-1)
y_train = y_train.unsqueeze(1)
```

## Define Neural Nets

```python
nu = 2
print("The real value: ", viscosity_coeff, " Our inital guess: ", nu)
```

```python
class NN(torch.nn.Module):
    def __init__(self):
        self.net = torch.nn.Sequential(
            torch.nn.Linear(2, 20),
            torch.nn.Tanh(),
            torch.nn.Linear(20, 30),
            torch.nn.Tanh(),
            torch.nn.Linear(30, 20),
            torch.nn.Tanh(),
            torch.nn.Linear(20, 20),
            torch.nn.Tanh(),
            torch.nn.Linear(20, 1),
            torch.nn.Tanh(),
        )
    def forward(self, x):
        out = self.net(x)
        return out
```

```python
class Net:
    def __init__(self):
        device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
        self.nu = torch.tensor([nu], requires_grad=True).float().to_device() 
        self.nu = torch.nn.Parameter(self.nu)
        self.model = NN().to(device)
        self.model.register_parameter("nu", self.nu)
        
        # meshgrid to perform PDE calc.
        self.h = 0.1 # dx
        self.k = 0.1 # dt
        x = torch.arange(-1, 1 + self.h, self.h) 
        t = torch.arange(0, 1 + self.k, self.k)

        self.X = torch.stack(torch.meshgrid(x, t)).reshape(2, -1).T
        
        ################# input the data ##########################
        self.X_train =  combined_tensor_x_train
        self.y_train =  y_train
        ###########################################################
        self.X = self.X.to(device)
        self.X.requires_grad = True
        self.X_train = self.X_train.to(device)
        self.y_train = self.y_train.to(device)

        # Adam optimizer
        self.adam = torch.optim.Adam(self.model.parameters())
        self.criterion = torch.nn.MSELoss()
        self.iter = 1
        
    def loss_func(self):
        self.adam.zero_grad()

        ################ stochastic loss ################################
        y_pred = self.model(self.X_train)
        loss_data = self.criterion(y_pred, self.y_train)
        #################################################
        
        # PDE Loss
        u = self.model(self.X)
        
        du_dX = torch.autograd.grad(
            u,
            self.X,
            grad_outputs = torch.ones_like(u),
            create_graph = True,
            retain_graph = True
        )[0]
        du_dt = du_dX[:, 1]
        du_dx = du_dX[:, 0]

        du_dXX = torch.autograd.grad(
            du_dX,
            self.X,
            grad_outputs = torch.ones_like(du_dX),
            create_graph = True,
            retain_graph = True
        )[0]
        du_dxx = du_dXX[:, 0]

        nu_calc = self.nu
        loss_pde = self.criterion(du_dt + 1*u.squeeze() * du_dx, nu_pde* du_dxx) # keep changing nu values to satisfy this PDE 
        
        loss = loss_pde + loss_data
        loss.backward()

        if self.iter % 100 == 0:
            print("iterations #:  ", self.iter, ". Loss value: ", loss.item(), "Real nu: ", viscosity_coeff, ". I-PINNs nu: ", self.nu.item())
        
        self.iter += 1
        return loss

    def train(self):
        self.model.train()

        for i in range(3e3):
            self.adam.step(self.loss_func)

    def eval_():
        self.model.eval()
```

```python
net = Net()
net.train()
net.model.eval()
```
